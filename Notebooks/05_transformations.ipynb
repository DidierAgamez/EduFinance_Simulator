```python
"""Importamos las librerias"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from utils.paths import DATA_DIR, FIG_DIR
from utils.loader import load_csv

SERIES_DATA_DIR = DATA_DIR / "time_series"
SERIES_DATA_DIR.mkdir(parents=True, exist_ok=True)

print(" DATA_DIR:", DATA_DIR)
print(" FIG_DIR:", FIG_DIR)
```

```python
"""Descargamos el dataset limpio"""

market_df_clean = load_csv(DATA_DIR / "clean_market_data.csv")
```

```python
# --- Usamos directamente el DataFrame limpio ---
prices_df = market_df_clean.copy() # tu DataFrame ya normalizado, sin nulos, fechas alineadas

#Pivotamos el dataframe
prices_pivot = prices_df.pivot(index='date', columns='ticker', values='close')

# --- Selecci贸n del DataFrame base ---
prices = prices_pivot.copy()
prices.index = pd.to_datetime(prices.index, errors="coerce")
prices = prices.sort_index()

# --- Aplicamos transformaciones ---

# Logaritmo natural de los precios
log_prices = np.log(prices.replace({0: np.nan}))

# Dfierenciacion de precios
diff_close = prices.diff().dropna()

# Diferenciaci贸n logar铆tmica
log_diff = log_prices.diff().dropna()

# Retornos simples
returns = prices.pct_change().dropna()

# Retornos al cuadrado
returns_sq = returns**2

# Media m贸vil (21 d铆as)
roll_mean_21 = prices.rolling(21).mean()

# Desviaci贸n est谩ndar m贸vil (21 d铆as)
roll_std_21 = prices.rolling(21).std()
```

```python
"""Guardamos los dataframes transformados"""

prices.to_csv(SERIES_DATA_DIR / "prices.csv")
diff_close.to_csv(SERIES_DATA_DIR / "diff_close.csv")
log_prices.to_csv(SERIES_DATA_DIR / "log_prices.csv")
log_diff.to_csv(SERIES_DATA_DIR / "log_diff.csv")
returns.to_csv(SERIES_DATA_DIR / "returns.csv")
returns_sq.to_csv(SERIES_DATA_DIR / "returns_sq.csv")   
roll_mean_21.to_csv(SERIES_DATA_DIR / "roll_mean_21.csv")
roll_std_21.to_csv(SERIES_DATA_DIR / "roll_std_21.csv")
```
